```python
import os
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# bring in CSV files
England = "England.csv"
Germany = "Germany.csv"
Italy = "Italy.csv"
Spain = "Spain.csv"

# Read in the data
England_data = pd.read_csv(England)
Germany_data = pd.read_csv(Germany)
Italy_data = pd.read_csv(Italy)
Spain_data = pd.read_csv(Spain)
```

```python
# merge the four datasets
frames = [England_data, Germany_data, Italy_data, Spain_data]
all_league_data = pd.concat(frames)
```

```python
all2_df = all_league_data.iloc[:, [0, 1, 2, 3, 4, 5, 6, 7]]
all2_df
```

```python
all4_df = all2_df.rename(columns = {"FTHG":"HomeScore", "FTAG":"AwayScore", "FTR":"Result"})

all4_df
```

```python
gcsv_path = "germany.csv"
icsv_path = "italy.csv"
scsv_path = "spain.csv"

ger_df = pd.read_csv(gcsv_path)
ita_df = pd.read_csv(icsv_path)
spa_df = pd.read_csv(scsv_path)
germany_df = ger_df.iloc[:, [0, 1, 2, 3, 4, 5, 6, 7]]
italy_df = ita_df.iloc[:, [0, 1, 2, 3, 4, 5, 6, 7]]
spain_df = spa_df.iloc[:, [0, 1, 2, 3, 4, 5, 6, 7]]
germany_df = germany_df.rename(columns = {"FTHG":"HomeScore", "FTAG":"AwayScore", "FTR":"Result"})
italy_df = italy_df.rename(columns = {"FTHG":"HomeScore", "FTAG":"AwayScore", "FTR":"Result"})
spain_df = spain_df.rename(columns = {"FTHG":"HomeScore", "FTAG":"AwayScore", "FTR":"Result"})
```

```python
# make lists by league for team IDs
BUN_teams = [
    "Augsburg",
    "Bayern Munich",
    "Dortmund",
    "Ein Frankfurt",
    "FC Koln",
    "Fortuna Dusseldorf",
    "Freiburg",
    "Hertha",
    "Hoffenheim",
    "Leverkusen",
    "M'gladbach",
    "Mainz",
    "Paderborn",
    "RB Leipzig",
    "Schalke 04",
    "Union Berlin",
    "Werder Bremen",
    "Wolfsburg"]
```

```python
SER_teams = [
    "Atalanta",
    "Bologna",
    "Brescia",
    "Cagliari",
    "Fiorentina",
    "Genoa",
    "Inter",
    "Juventus",
    "Lazio",
    "Lecce",
    "Milan",
    "Napoli",
    "Parma",
    "Roma",
    "Sampdoria",
    "Sassuolo",
    "Spal",
    "Torino",
    "Udinese",
    "Verona"]
```

```python
LIG_teams = [
    "Alaves",
    "Ath Bilbao",
    "Ath Madrid",
    "Barcelona",
    "Betis",
    "Celta",
    "Eibar",
    "Espanol",
    "Getafe",
    "Grenada",
    "Leganes",
    "Levante",
    "Mallorca",
    "Osasuna",
    "Real Madrid",
    "Sevilla",
    "Sociedad",
    "Valencia",
    "Valladolid",
    "Villarreal"]
```

```python
cols = ["HomeTeam ID", "AwayTeam ID", "City"]

BUNhid_df = pd.DataFrame(columns=cols)
BUNaid_df = pd.DataFrame(columns=cols)

for team in BUN_teams:
    
    germany_df["HomeTeam"] = team
    BUNhid_df["HomeTeam ID"] = "BUN"+str(BUN_teams.index(team))
    germany_df["AwayTeam"] = team
    BUNaid_df["AwayTeam ID"] = "BUN"+str(BUN_teams.index(team))
    BUNhid_df = germany_df.append(BUNhid_df[cols], ignore_index=True)
    BUNaid_df = germany_df.append(BUNaid_df[cols], ignore_index=True)
    
frames = [BUNhid_df, BUNaid_df]

BUNallid_df = pd.concat(frames)
```

```python
SERhid_df = pd.DataFrame(columns=cols)
SERaid_df = pd.DataFrame(columns=cols)

for team in SER_teams:
    
    italy_df["HomeTeam"] = team
    SERhid_df["HomeTeam ID"] = "SER"+str(SER_teams.index(team))
    italy_df["AwayTeam"] = team
    SERaid_df["AwayTeam ID"] = "SER"+str(SER_teams.index(team))
    SERhid_df = italy_df.append(SERhid_df[cols], ignore_index=True)
    SERaid_df = italy_df.append(SERaid_df[cols], ignore_index=True)
    
frames = [SERhid_df, SERaid_df]

SERallid_df = pd.concat(frames)
```

```python
cols = ["HomeTeam ID", "AwayTeam ID", "City"]

LIGhid_df = pd.DataFrame(columns=cols)
LIGaid_df = pd.DataFrame(columns=cols)

for team in LIG_teams:
    
    spain_df["HomeTeam"] = team
    LIGhid_df["HomeTeam ID"] = "LIG"+str(LIG_teams.index(team))
    spain_df["AwayTeam"] = team
    LIGaid_df["AwayTeam ID"] = "LIG"+str(LIG_teams.index(team))
    LIGhid_df = spain_df.append(LIGhid_df[cols], ignore_index=True)
    LIGaid_df = spain_df.append(LIGaid_df[cols], ignore_index=True)
    
frames = [LIGhid_df, LIGaid_df]

LIGallid_df = pd.concat(frames)
```

```python
BUNallid_df.to_csv("Germany_teamID.csv")
SERallid_df.to_csv("Italy_teamID.csv")
LIGallid_df.to_csv("Spain_teamID.csv")
```

```python
import pandas as pd
import os
import requests
import json
import matplotlib.pyplot as plt
```

```python
EPL_df = pd.read_csv("../final-resources/EPL_scores_att_weather.csv")
Bundesliga_df = pd.read_csv("../final-resources/Bundesliga_scores_weather.csv")
LaLiga_df =  pd.read_csv("../final-resources/la_liga_scores_weather.csv")
SerieA_df = pd.read_csv("../final-resources/Serie_A_weather.csv")
```

```python
vertical_stack = pd.concat([EPL_df,Bundesliga_df,LaLiga_df,SerieA_df], axis=0)
vertical_stack
```

```python
vertical_stack
all_weather = vertical_stack[["Div","Date","HomeScore","AwayScore","Result","Avg Temp (in F)"]]
all_weather
all_weather.to_csv("../final-resources/All_Weather_Data.csv")
colors = all_weather["Result"].replace(["H","A","D"], ["b","b","b"])
#Scatter plot
xval = all_weather["Avg Temp (in F)"]
yval = [all_weather["HomeScore"]]

plt.scatter(xval, yval[0], c=colors,label="Home")
plt.figure(figsize = (8,8))

#ax1.set_title("Weather vs. Score")
#ax.legend()
#plt.show()
#plt.savefig("Fig1.png")
colors = all_weather["Result"].replace(["H","A","D"], ["r","b","g"])
#Scatter plot
xval = all_weather["Avg Temp (in F)"]
yval = [all_weather["AwayScore"]]

plt.scatter(xval, yval[0], c=colors,label="Home")
plt.figure(figsize = (8,8))
```
